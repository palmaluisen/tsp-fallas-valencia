# Get the links from url 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Configurar el WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Ejecutar en modo headless
driver = webdriver.Chrome(options=options)

# Abrir la página
url = "https://www.valenciabonita.es/2025/03/14/mapa-fallas-2025/"
driver.get(url)

# Buscar los enlaces dentro del cuerpo de la página
fallas = driver.find_elements(By.XPATH, "//a[contains(text(), 'Falla ')]")

# Extraer nombres y enlaces
dict_fallas = {}
for falla in fallas:
    nombre = falla.text
    enlace = falla.get_attribute("href")
    if "maps.app" in enlace:
        print(f"{nombre}: {enlace}")
        dict_fallas[nombre] = enlace

# Cerrar el navegador
driver.quit()
```

# Get long_url

```python
import requests

def expandir_url(short_url):
    response = requests.head(short_url, allow_redirects=True)
    return response.url

for nombre, enlace in dict_fallas.items():
    long_url = expandir_url(enlace)
    print(f"{nombre}: {long_url}")
    dict_fallas[f"{nombre}"] = long_url
```

# Get the coordinates

```python
import pandas as pd
df = pd.DataFrame(dict_fallas.items(), columns=["Falla", "Enlace"])
df
```

```python
# Extraer latitud y longitud
df["lat"] = df["Enlace"].apply(lambda x: x.split("@")[1].split(",")[0])
df["lon"] = df["Enlace"].apply(lambda x: x.split("@")[1].split(",")[1])
```
